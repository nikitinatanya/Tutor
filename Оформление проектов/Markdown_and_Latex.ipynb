## Markdown

*Markdown* - это облегченный язык разметки, который можно использовать для добавления элементов форматирования в текстовые документы открытого текста.

Чтобы создать разрыв строки или новую строку `<br>`, необходимо завершить строку двумя или более пробелами, а затем ввести return.

### Списки

#### Нумерованный список  


Вне зависимости от порядка нумерации, порядок вывода будет корректен 

1. Фактор 1  
2. Фактор 2
3. Фактор 3
5. Фактор 4

#### Маркированный список  


Возможно использовать в качестве символа `-`, `*`, `+`

- Фактор 1
- Фактор 2  
- Фактор 3

#### Вложенный список  

Вложенные пункты создаются четырьмя пробелами перед маркером пункта.

1. Фактор 1
    - позиция 1.1  
    - позиция 1.2  
2. Фактор 2  
    - позиция 2.1  
        - предмет 2.1.1
        - предмет 2.1.2
    - позиция 2.2

### Ссылки

Чтобы создать ссылку, заключите текст ссылки в скобки (например, [ЯндексПрактикум]), а затем сразу же следуйте за ней с URL-адресом в скобках (например, (https://practicum.yandex.ru)). 

Мне нравится комьюнити **[ЯндексПрактикум](https://practicum.yandex.ru)**   

### Цитаты

Цитаты можно оформить с помощью символа `>`.  

>Без труда не вынешь рыбку из пруда.


### Изображения

Чтобы добавить изображение, добавьте восклицательный знак (!), за которым следует альтернативный текст в скобках и путь или URL-адрес к ресурсу изображения в скобках. 

## LaTeX

LaTeX — язык разметки и система верстки, предназначенная для набора научных текстов. В рамках курса может потребоваться для написания единиц измерения и несложных формул.

Для написания формулы в тексте обрамляем ее `$`  

*Средняя площадь квартиры - $25м^2$*

Для написания формулы на отдельной строке используем `$$`  

Расчет точки безубыточности осуществляется по следующей формуле:

$$Q_б = \frac{FC}{P-AVC}\$$

$Q_б$ - объем выпуска  
$FC$ - постоянные издержки  
$P$ - цена  
$AVC$ - средние переменные издержки
