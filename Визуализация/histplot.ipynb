```python
import pandas as pd
import math as mth
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
pd.set_option('display.float_format', '{:,.2f}'.format)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 10)
pd.set_option('display.max_colwidth', 1000)
```

```python
sns.set(font_scale = 1.5, style="darkgrid")
from matplotlib.colors import ListedColormap
cmap = ListedColormap(sns.color_palette())
```

```python
# загрузим датасеты
users = pd.read_csv('https://stepik.org/media/attachments/lesson/426730/active_users.csv')
groups = pd.read_csv('https://stepik.org/media/attachments/lesson/426730/groups.csv')
purchases = pd.read_csv('https://stepik.org/media/attachments/lesson/426730/purchases.csv')
```

```python
# объединие файлов
test_group = users.merge(groups, how = 'left', on = 'user_id')
revenue_test_group = purchases.merge(groups, how = 'left', on = 'user_id')
```

## Распределение выручки в группах тестирования

### В одних осях

```python
fig, ax = plt.subplots(figsize=(18, 8))                                                                # создание "рамки"
ax = sns.histplot(x = revenue_test_group.query('group == "B"').revenue, kde = True, label = 'группа B')                    # распределение выручки группа В
sns.histplot(x = revenue_test_group.query('group == "A"').revenue, kde = True, color = 'green', label = 'группа A', ax=ax) # распределение выручки группа А
plt.suptitle('Распределение выручки в группах тестирования',fontsize = 27)                             # наименование графика
plt.legend()                                                                                           # легенда
plt.ylabel('Количество пользователей')                                                                 # замена подписи оси y
plt.xlabel('Выручка')                                                                                  # замена подписи оси x
plt.tight_layout()
plt.show()                                                                                             # избавимся от системного сообщения перед графиком

```

## С применением subplot()

```python
fig, ax = plt.subplots(1, 2, figsize = (18, 8))
sns.histplot(x = revenue_test_group.query('group == "A"').revenue, kde = True, label = 'группа A', ax=ax[0]).set(
    ylabel = 'Количество пользователей чел.', xlabel = 'Выручка', yticks =[20,40,60,80,100,120,140]
)
sns.histplot(x = revenue_test_group.query('group == "B"').revenue, kde = True, label = 'группа A', ax=ax[1]).set(
    ylabel = 'Количество пользователей чел.', xlabel = 'Выручка', yticks = [20,40,60,80,100,120,140]
)

plt.suptitle('Распределение выручки в группах тестирования', fontsize=27)

plt.tight_layout()
plt.show()

```
